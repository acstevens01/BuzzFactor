```python
import pandas as pd
def get_direct_link(shared_link):
    return 'https://drive.google.com/uc?export=download&id=' + shared_link.split('/')[-2]

link = "https://drive.google.com/file/d/1tMN5OCbXtX9st-Q5oDaCqQ7nsZ5Iw7ma/view?usp=drive_link"

direct_link = get_direct_link(link)
players = pd.read_csv(direct_link)
players.columns = ['original_names']
players['search_query'] = players['original_names'] + " UFC wikipedia"
```

```python
import requests
from bs4 import BeautifulSoup
import urllib.parse

def get_first_search_result_title(query):
    query = urllib.parse.quote_plus(query)
    url = f"https://www.google.com/search?q={query}"

    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36",
        'Accept-Language': 'en-US,en;q=0.5'
    }

    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.text, "html.parser")
    first_result = soup.find("div", {"class": "yuRUbf"})
    if first_result:
        title = first_result.find("h3").text
        
        link = first_result.find("a")['href']
        wikipedia_url = urllib.parse.unquote(link)

        if 'wikipedia.org' in wikipedia_url:
            parsed_url = urllib.parse.urlparse(wikipedia_url)
            wiki_title = parsed_url.path.split('/')[-1].replace('_', ' ')
        else:
            wiki_title = "No Wikipedia page"

        return wiki_title

    return None

count = 0
for index, row in players.iterrows():
    if index < count:
        continue
    query = row["search_query"]
    result_title = get_first_search_result_title(query)
    if not result_title:
        print(count,index,row['original_names'])
        break
    players.at[index, "wiki_title"] = result_title if result_title else "No search results found"
    print(index)
```

```python
not_found_names = []

for index, row in players.iterrows():
    wiki_title = row['wiki_title']
    if wiki_title in ["No Wikipedia page", "No search results found"]:
        not_found_names.append(row['original_names'])

print("Original Names with No Wikipedia page or No search results:")
print(not_found_names, len(not_found_names))
```

```python
players.to_csv("ufc_wiki_fighters.csv", encoding='utf-8', index=False)
```
